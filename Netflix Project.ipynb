# movies aur TV shows ka count chart

```python
import matplotlib.pyplot as plt
type_count = df['type'].value_counts()
type_count.plot(kind='bar')
plt.title('Movies Vs TV shows on netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
country_count = df['country'].value_counts().head(10)
country_count.plot(kind='bar')
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Number of Shows')
plt.show()
```

## Release year trend chart

```python
year_count = df['release_year'].value_counts().sort_index()
year_count
```

```python
year_count.plot(kind='line')
plt.title("Netflix Content Release Trend")
plt.xlabel("Year")
plt.ylabel("Number of Shows")
plt.show()
```

### Insights
#### Netflix content 2015 ke baad repidly growth hua hai 
#### Streaming industry growth dikhata hai
